### Data Transformation

```python
def convert_gender(row):
    if row['gender'] == 1:
        return 'male'
    elif row['gender'] ==2:
        return 'female'
    else:
        return 'unknown'
    
#2.apply function
df_bike.apply(convert_gender, axis=1)
```

```python
df_bike['jenis_kelamin'] = df_bike.apply(convert_gender, axis=1)
```

```python
df_bike
```

```python
def route(row):
    return row['start station name']+' - '+row['end station name']

df_bike['route'] = df_bike.apply(route,axis=1)
df_bike[['start station name','end station name','route']]
```

```python
df_bike['tripduration (minute)']= df_bike['tripduration']/60
```

```python
df_bike[['tripduration','tripduration (minute)']]
```

```python
def to_minute(row):
    return str(round(row['tripduration']/60))+" menit"

df_bike['tripduration (minute) func'] = df_bike.apply(to_minute,axis=1)
```

```python
df_bike[['tripduration','tripduration (minute)','tripduration (minute) func']]
```
