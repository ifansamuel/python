```
# Latihan

# Buat sebuah function untuk menentukan kata mana yang memiliki jumlah huruf paling banyak
# Input: 2 string
# Output: String yang memiliki paling banyak huruf, kalau jumlah hurufnya sama kalian return "sama"

# Input: "remot" "televisi"
# Output: "televisi 8"

# Input: "gitar" "remot"
# Output: "sama"

def perbandingan_kata(string1, string2):
  if len(string1) > len(string2):
    return string1 + " " + str(len(string1))

  elif len(string1) < len(string2):
    return string2 + " " + str(len(string2))
  
  elif len(string1) == len(string2):
    return "sama"

a = "televisi"
b = "remot"

perbandingan_kata(a, b)
```

```
# Latihan

# Coba kalian buat function untuk menghapus element dari list of strings --> ["apel", "jeruk", "pisang"]
# Input: parameter 1 = string  ,  parameter 2 = list
# Output: list setelah element dihapus

# Input: 3 , [1, 2, 3, 4, 5]
# Output: [1, 2, 4, 5]

def menghapus_element(value, list):
  list.remove(value)
  return list

menghapus_element(2, [1, 2, 3, 4, 5])
```

```
# Latihan

# Coba kalian buat function untuk menghitung jumlah huruf pada list of strings
# Input: 1 list of strings --> ["aku", "adalah", "anak", "gembala"]
# Output: jumlah huruf pada list tersebut --> "4 kata 20 huruf"

kalimat = ["adalah", "anak", "gembala"]

def hitung(list):
  total = 0
  kata = len(list)
  for i in kalimat:
    total += len(i)
  return "{} kata {} huruf".format(kata, total)

hitung(kalimat)
```
