## Data Analisys - Analyzing Booking Hotel Data

Melakukan analisis terhadap data booking hotel, untuk melihat tren dari waktu ke waktu.

```python
hotel = pd.read_csv('http://bit.ly/dwp-data-hotels')
```

```python
hotel.head()
```

#### Questions

1. Berapa total nights dan total revenue di masing-masing hotel di tiap bulan?
2. Berapa total bookings pada setiap market segment di tiap bulan?
3. Apakah ada hubungan antara lead time dan cancellation?
4. Berapa banyak cancellation pada setiap market segment?

#### Wrangle

```python
# Cek null data, dan misalkan di kolom yang akan kalian analisis itu ada yg NULL
# kalian bisa isi data null tersebut

# 1. Tambahkan kolom "total_nights" --> penjumlahan dari "stays_in_weekend_nights" dan "stays_in_week_nights"
hotel['total_nights'] = hotel['stays_in_weekend_nights'] + hotel['stays_in_week_nights']

# 2. Tambahkan kolom "revenue" --> perkalian dari "adr" dan "total_nights"
hotel['revenue'] = hotel['adr'] * hotel['total_nights']

# 3. Tambahkan kolom "arrival_month" --> diambil dari "arrival_date" dengan format "2017-01"
hotel['arrival_date'] = pd.to_datetime(hotel['arrival_date'])
hotel['arrival_month'] = hotel['arrival_date'].dt.strftime('%Y-%m')

# 4. Tambahkan kolom untuk mengelompokan lead_time

def group_lead_time(row):
    if row['lead_time'] <= 200:
        return '0-200'
    elif row['lead_time'] <= 400:
        return '201-400'
    elif row['lead_time'] <= 600:
        return '401-600'
    else:
        return '600>'
    
hotel['lead_time_category'] = hotel.apply(group_lead_time, axis=1)
```

#### Explore

#### 1. Total Nights dan Total Revenue tiap bulan di setiap Hotel

```python
# Jangan lupa untuk filter datanya yang memiliki is_canceled = 0
hotel_filtered = hotel[hotel['is_canceled'] == 0]
```

```python
# Aggregasi Data

# - Yang dicari (bulan - hotel)
# - Yang dihitung (total_nights)
# - Fungsi aggregasi (sum)
```

```python
# Visualisasi Menggunakan Seaborn - Total Nights per Month

night_per_month = hotel_filtered.groupby(['arrival_month', 'hotel'], as_index=False)['total_nights'].sum()

plt.figure(figsize=(15,5))
sns.lineplot(data=night_per_month, x='arrival_month', y='total_nights', hue='hotel')
```

```python
# Visualisasi Menggunakan Pandas - Total Revenue per Month

night_per_month = hotel_filtered.groupby(['arrival_month', 'hotel'])['revenue'].sum().unstack()

night_per_month.plot.line(figsize=(15,5))
```

#### 2. Total Bookings di setiap market segments di tiap bulan

```python
hotel.head(1)
```

```python
# Aggregasi Data

# - Yang dicari (bulan - market_segments)
# - Yang dihitung ()
# - Fungsi aggregasi (count)
```

```python
# Visualisasi menggunakan seaborn

total_segment = hotel_filtered.groupby(['arrival_month', 'market_segment'], as_index=False)['adr'].count()

plt.figure(figsize=(15,5))
sns.barplot(data=total_segment, x='arrival_month', y='adr', hue='market_segment')
```

```python
# Visualisasi menggunakan pandas - stacked bar

total_segment = hotel_filtered.groupby(['arrival_month', 'market_segment'])['adr'].count().unstack()

total_segment.plot.bar(figsize=(15,5), stacked=True)
```

```python
data = hotel_filtered.groupby(['arrival_month', 'market_segment'])['adr'].count().unstack()

data.plot.bar(stacked=True)
#data.plot.bar(stacked=True)
```

```python
# Visualisasi menggunakan pandas - stacked bar percentage

total_segment = hotel_filtered.groupby(['arrival_month', 'market_segment'])['adr'] \
                              .count() \
                              .groupby(level=0).apply(lambda x: 100 * x / float(x.sum())) \
                              .unstack()

total_segment.plot.bar(figsize=(15,5), stacked=True, colormap='Blues')
```

#### 3. Cancellation dilihat dari Lead Time

```python
# 1. Kelompokan "lead_time" berdasarkan range data nya (buat 1 kolom baru)
# 2. Rule:
#    - < 200     --> '0-200'
#    - 201 - 400 --> '201-400'
#    - 401 - 600 --> '401-600'
#    - 600 >     --> '600>'
```

```python
cancel_lead_time = hotel.groupby(['lead_time_category', 'is_canceled'])['adr'].count()\
                        .groupby(level=0).apply(lambda x: 100 * x / float(x.sum())) \
                        .unstack()

cancel_lead_time.plot.bar(stacked=True)
```

#### 4. Cancellation dilihat dari Market Segment

```python
cancel_lead_time = hotel.groupby(['market_segment', 'is_canceled'])['adr'].count()\
                        .groupby(level=0).apply(lambda x: 100 * x / float(x.sum())) \
                        .unstack()

cancel_lead_time.plot.bar(stacked=True)
```
