```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pandasql
```

```python
data_bike = pd.read_csv('http://bit.ly/dwp-data-bike')
data_sales = pd.read_csv('http://bit.ly/dwp-data-ecommerce')
data_movies = pd.read_csv('http://bit.ly/dwp-data-movies-sample')
data_ratings = pd.read_csv('http://bit.ly/dwp-data-ratings-sample')
```

```python
df_movies = data_movies
df_ratings = data_ratings
df_bike = data_bike
df_sales = data_sales
```

Latihan
cari rata tripduration dari setiap usertype

```python
query = """
        SELECT usertype,
            AVG(tripduration) as avg
        FROM df_bike
        WHERE tripduration > 300 
            AND gender = 1
        GROUP BY usertype
        """
result = pandasql.sqldf(query)
result
```

```python
df_bike_filtered = df_bike[(df_bike['gender']==1) & (df_bike['tripduration']>300)]
df_bike_filtered.groupby('usertype',as_index = False)['tripduration'].mean()
```

latihan cari 5 rute dengan total trip duration paling banyak

```python
query = """
        SELECT `start station name`,
                `end station name`,
                sum(tripduration) as total
        FROM df_bike
        GROUP BY `start station name`,
                `end station name`
        ORDER BY total desc
        LIMIT 5
        """
result = pandasql.sqldf(query)
result
```

### Question

1. Total produk terjual setiap bulan itu berapa dan dipisahkan setiap kategori
2. Jumlah transaksi disetiap jam dan tiap hari ada berapa

### Wrangle

1

```python
query = """
        SELECT strftime('%Y-%m', order_datetime) as date,
                product_category,
                count(*) as total,
                sum(product_cost+delivery_cost) as cost
        FROM df_sales
        GROUP BY date, 
            product_category
        ORDER BY date asc,
            product_category asc
        """
result1 = pandasql.sqldf(query)
result1.head()
```

```python
plt.figure(figsize=(15,4))
sns.lineplot(data=result1, x='date', y='total', hue='product_category')

plt.show()
```

2

```python
query = """
        SELECT strftime("%w", order_datetime) as day,
                strftime("%H", order_datetime) as hour,
                count(*) as total
        FROM df_sales
        GROUP BY day,hour
        ORDER BY day asc, hour asc
        """
result2 = pandasql.sqldf(query)
result2
```

```python
plt.figure(figsize=(15,4))
sns.barplot(data=result2,x='hour',y='total',hue='day')
plt.show()
```

#### Pake Pandas

```python
df_sales['total_cost']=df_sales['product_cost']+ df_sales['delivery_cost']
df_sales['order_datetime'] = pd.to_datetime(df_sales['order_datetime'])
df_sales['order_month']=df_sales['order_datetime'].dt.strftime('%Y-%m')
df_sales['order_hour']=df_sales['order_datetime'].dt.strftime('%H')
df_sales['order_day']=df_sales['order_datetime'].dt.strftime('%A')
df_sales['order_day']=pd.Categorical(df_sales['order_day'],
                                    ordered=True,
                                    categories=['Monday',
                                                   'Tuesday',
                                                   'Wednesday',
                                                   'Thursday',
                                                   'Friday',
                                                   'Saturday',
                                                   'Sunday'])
```

```python
query= """
        SELECT 
            order_month,
            product_category,
            count(*) as quantity,
            sum(total_cost) as total    
        FROM 
            df_sales
        GROUP BY 
            order_month,
            product_category
        
        """
result3 = pandasql.sqldf(query)
result3
```

```python
plt.figure(figsize=(15,4))
sns.lineplot(data=result3, x='order_month', y='total', hue='product_category')

plt.show()
```

```python
query= """
        SELECT 
            order_day,
            order_hour,
            count(*) as total    
        FROM 
            df_sales
        GROUP BY 
            order_day,
            order_hour
        
        """
result4 = pandasql.sqldf(query)
result4
#ga urut
```

```python
result4_2 = df_sales.groupby(['order_day','order_hour'],as_index=False)['order_id'].count()
```

```python
plt.figure(figsize=(14,5))
sns.barplot(data=result4_2,x='order_day',y='order_id',hue='order_hour')
plt.xlabel('Hari')
plt.ylabel('Total')
plt.title('Day-Hour Total Sold Product')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
